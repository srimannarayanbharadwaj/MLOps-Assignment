```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

DATA_PATH = Path('../data/processed/heart_cleveland_clean.csv')
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.shape, df.isna().sum()
```

## Target Encoding

`target = 0` means no heart disease. Original UCI labels `1`, `2`, `3`, and `4` are collapsed to `1` for disease present.

```python
df['target'].value_counts().sort_index()
```

```python
numeric_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df[numeric_features].hist(bins=20, figsize=(12, 8), color='#4c78a8', edgecolor='white')
plt.suptitle('Numeric Feature Distributions', y=1.02)
plt.tight_layout()
plt.show()
```

Write your observation here: Which feature distributions look skewed, clustered, or clinically interesting?

```python
plt.figure(figsize=(11, 9))
sns.heatmap(df.corr(numeric_only=True), cmap='vlag', center=0, linewidths=0.4)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

Write your observation here: Which features appear most related to the target? Are any predictors strongly related to each other?

```python
class_counts = df['target'].map({0: 'No disease', 1: 'Disease'}).value_counts()
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Target Class Balance')
plt.xlabel('Target class')
plt.ylabel('Patient count')
plt.show()
```
